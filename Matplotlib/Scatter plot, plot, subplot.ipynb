## MatplotLib Tutorial

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
## Simple Examples
x=np.arange(0,10)
y=np.arange(11,21)
x
```

**scatter Plot**

The scatter() function plots one dot for each observation. It needs two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis:

```python
##plotting using matplotlib 
                                        ##plt scatter plot
plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('Test.png')
```

## By default, the plot() function draws a line from point to point.

```python
y=x*x
```

```python
## plt plot
plt.plot(x,y,'g*',linestyle='dashed',linewidth=2, markersize=12) 
                        # g* means green color and each point is indicated by *
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2d Diagram')
```

## ## Creating Subplots

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'go')
```
