## Logical Operators

```python
# Bitwise  AND operator
a=5   #0000 0101
b=6   #0000 0110
c=a & b
c
```

```python
# Bitwise  OR operator
d = a | b
d
```

```python
# Bitwise  NOT operator
k=15   #0000 1111
E=~k   #1111 0000 --> 2'complement --> 0001 0000 = -16
E
```

```python
# Bitwise  right shift operator
f=5 #0000 0101
f>>2
```

```python
# Bitwise  left shift operator
f1=5 #0000 0101
f1<<2
```
