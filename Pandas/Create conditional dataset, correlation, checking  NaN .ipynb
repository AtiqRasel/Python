## Check NaN values

```python
data = {'first_set': [1,2,3,4,5,np.nan,6,7,np.nan,np.nan],
        'second_set': ['a','b',np.nan,np.nan,'c','d','e',np.nan,np.nan,'f'],
        'third_set':['aa',np.nan,'bb','cc',np.nan,np.nan,'dd',np.nan,np.nan,'ee']
        }

df = pd.DataFrame(data,columns=['first_set','second_set','third_set'])

print (df)
```

## Count NaN values under a single DataFrame column:

```python
df['first_set'].isna().sum()
```

## Count columnwise NaN values under an entire DataFrame:

```python
df.isna().sum()
```

## Count NaN values under an entire DataFrame:

```python
df.isna().sum().sum()
```

## Count NaN values across a single DataFrame row:

```python
df.loc[[0,1,2]].isna().sum().sum()
```
