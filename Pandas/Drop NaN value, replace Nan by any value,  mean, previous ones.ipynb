## Pandas

```python
## First step is to import pandas

import pandas as pd
import numpy as np
```

## Remove NaN values

```python
def dataset():
    data = {'first_set': [1,2,3,4,5,np.nan,6,7,np.nan,np.nan],
            'second_set': [1.5,4.8,np.nan,np.nan,4.3,7.3,6.1,np.nan,np.nan,8.5],
            'third_set':[3.4,np.nan,4.3,5.3,np.nan,np.nan,7.9,np.nan,np.nan,9.3]
            }
    df = pd.DataFrame(data,columns=['first_set','second_set','third_set'])
    return df
```

```python
df=dataset()
df
```

## Check null values in the entire dataframe

```python
df.isna().sum()
```

## Replace NaN value of a column by its mean value

```python
import math
meanTestScore1=math.floor(df['first_set'].mean())
meanTestScore2=math.floor(df['second_set'].mean())
meanTestScore3=math.floor(df['third_set'].mean())
print(meanTestScore1, meanTestScore2, meanTestScore3)
```

```python
df['first_set']=df['first_set'].fillna(meanTestScore1)
df['second_set']=df['second_set'].fillna(meanTestScore2)
df['third_set']=df['third_set'].fillna(meanTestScore3)
df
```

```python
import seaborn as sns
sns.heatmap(df.isnull(), cmap='viridis')
```
